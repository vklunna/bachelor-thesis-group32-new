```python
# Load saved results
summary_df = pd.read_csv("../2_output/grid_search_results.csv")
with open("../2_output/best_weights.json", "r") as f:
    best_weights = json.load(f)

# Rerun extraction using best weights
extracted_df = test_weights_on_pdfs("../0_data/table_extraction_accuracy2.xlsx", best_weights)

# Load original labels
labels_df = pd.read_excel("../0_data/table_extraction_accuracy2.xlsx")

# === Comparison Logic ===

# Rename old extracted columns
labels_df_renamed = labels_df.rename(columns={
    "extracted start": "extracted_start_old",
    "extracted end": "extracted_end_old"
})

# Rename new extracted columns
extracted_df_renamed = extracted_df.rename(columns={
    "extracted_start": "extracted_start_new",
    "extracted_end": "extracted_end_new"
})

# Merge both on pdf_path
comparison_df = labels_df_renamed.merge(
    extracted_df_renamed,
    on="pdf_path",
    how="inner"
)

# Compare values
comparison_df["start_changed"] = comparison_df["extracted_start_old"] != comparison_df["extracted_start_new"]
comparison_df["end_changed"] = comparison_df["extracted_end_old"] != comparison_df["extracted_end_new"]

# Filter differences
changed_df = comparison_df[
    comparison_df["start_changed"] | comparison_df["end_changed"]
].copy()

# Show changes
print(f"🔍 Total files with different start/end: {len(changed_df)}")
display(changed_df[[
    "pdf_path",
    "extracted_start_old", "extracted_start_new",
    "extracted_end_old", "extracted_end_new"
]])
```
